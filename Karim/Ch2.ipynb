To calculate the lk norm of a vector, I can use the SciPy library.
Check this: https://www.digitalocean.com/community/tutorials/norm-of-vector-python

```python
import numpy as np
```

```python
arr=np.array([2,-5,4])
arr
```

L0 rerturns the number of non zero elements in a vector. Cardinality.

```python
np.linalg.norm(arr,0)
```

L1 is the Manhattan distance. It is the orthogonal distance. It reurns an absolute value sum of all elements.<br> It is related to the Mean Absolute Error (MAE).

```python
np.linalg.norm(arr,1)
```

L2 is the Eucledian distance. It is the square root of the sum of the absolut value of each element. <br> It is related to Root Mean Square Error (RMSE)

```python
np.linalg.norm(arr,2)
```

LInfinity give the maximum absolute value in the vector. It gets the largest number (in absolute value).

```python
np.linalg.norm(arr,np.Inf)
```

```python
from pathlib import Path
import pandas as pd
import tarfile
import urllib.request
```

```python
def load_housing_data():
    tarball_path=Path("datasets/housing.tgz")
    if not tarball_path.is_file():
        Path("datasets").mkdir(parents=True,exist_ok=True)
        url="https://github.com/ageron/data/raw/main/housing.tgz"
        urllib.request.urlretrieve(url,tarball_path)
        with tarfile.open(tarball_path) as housing_tarball:
            housing_tarball.extractall(path="datasets")
        return pd.read_csv("datasets/housing/housing.csv")
```

```python
housing=load_housing_data()
```
